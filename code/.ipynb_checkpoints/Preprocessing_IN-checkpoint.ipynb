```python
import pandas as pd
import numpy as np
```

```python
station_info = pd.read_csv("/scratch/brown/train52/data/gsod.stn.csv")
station_IN = station_info['state'] == 'IN'
station_IN = station_info[station_IN]
print(station_IN.columns)
station_IN.to_csv("/scratch/brown/train52/data/gsod.stn_onlyIN.csv",index=False)
```

```python
k = station_IN['station_id'] == 724370
station_IN[k]
```

```python
final= pd.DataFrame(columns = station_IN.columns)
nrows=5e7
for ob_data in pd.read_csv("/scratch/brown/train52/data/gsod.obs.csv",chunksize=nrows):
    print(ob_data.columns)
    joined_df = station_IN.merge(ob_data, on = ['station_id','wban_id'],how='inner')
    print("joined ",joined_df.shape)
    final = final.append(joined_df,ignore_index=True)
    print("final ",final.shape)
```

```python
final.to_csv("/scratch/brown/train52/data/obs_IN.csv",index=False)
```
